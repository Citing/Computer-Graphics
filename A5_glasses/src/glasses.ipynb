```python
import cv2
from matplotlib import pyplot as plt
```

```python
#display color histogram
def showHistogram(img):
    color = ('b', 'g', 'r')
    for i, color in enumerate(color):
        hist = cv.calcHist([img], [i], None, [256], [0, 256])
        plt.plot(hist, color)
        plt.xlim([0, 256])
    plt.show()
```

```python
fileName = '3.png'
imgSrc = cv.imread(fileName)
plt.imshow(imgSrc[:,:,::-1]) #convert BGR to RGB
plt.xticks([]), plt.yticks([]) #hide the axises
plt.show() #display the image
```

```python
showHistogram(imgSrc)
```

```python
# For a person with weak red cone cells
b, g, r = cv2.split(imgSrc)
for i in range(len(r)):
    r[i] = r[i] * 0.4
img_redWeak = cv2.merge([r, g, b])
plt.imshow(img_redWeak) #convert BGR to RGB
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
showHistogram(img_redWeak)
```

```python
# He wears a pair glasses filtering out green light.
for i in range(len(g)):
    g[i] = g[i] * 0.3
img_glasses = cv2.merge([r, g, b])
plt.imshow(img_glasses) #convert BGR to RGB
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
showHistogram(img_glasses)
```
